```python
import pandas as pd
import plotly.express as px
def read_results(strategy = 'random'):
    return pd.read_parquet(f"../data/model_results/{strategy}/_parquet/")

results = read_results()
results.loc[:,['steps', 'budget', 'initial_budget']].melt(id_vars='initial_budget')
px.line(results.loc[:,['steps', 'budget', 'initial_budget']].melt(id_vars=['initial_budget', 'steps']),
                       x="steps", 
                       y="value", 
                       color='initial_budget')
```
